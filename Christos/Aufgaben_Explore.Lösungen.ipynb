```python
salary_currency_count =df.loc[:, 'salary_currency'].unique()
print(salary_currency_count)
for index, row in df.iterrows():
    if row['salary_currency'] == 'EUR':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary']
    elif row['salary_currency'] == 'USD':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 1.07
    elif row['salary_currency'] == 'GBP':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] / 1.15
    elif row['salary_currency'] == 'HUF':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.00270
    elif row['salary_currency'] == 'INR':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.01127
    elif row['salary_currency'] == 'JPY':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.00668
    elif row['salary_currency'] == 'CNY':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.13
    elif row['salary_currency'] == 'MXN':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.05
    elif row['salary_currency'] == 'CAD':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.69645
    elif row['salary_currency'] == 'DKK':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.13423
    elif row['salary_currency'] == 'PLN':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.22314
    elif row['salary_currency'] == 'SGD':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.69284
    elif row['salary_currency'] == 'CLP':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.00117
    elif row['salary_currency'] == 'BRL':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.18978
    elif row['salary_currency'] == 'TRY':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.04338
    elif row['salary_currency'] == 'AUD':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] * 0.62204
    elif row['salary_currency'] == 'CHF':
        df.loc[index, 'salary_in_euro'] = df.loc[index, 'salary'] / 1.03040
    else:
        print('Currency not known')
df_new = df.loc[:,'salary_in_euro']
df_new
pd.options.display.float_format = '{:,.2f}'.format
df_new
df.sort_values(by=['salary_in_euro'], ascending=False)


```
